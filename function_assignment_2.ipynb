```python
# Q.9 Countdown: Write a function that takes a number x and uses a while loop to count down to zero.

def count_down(x):
    while x >= 0:
        print(x)
        x -= 1
```

```python
# Q.12 Simple Calculator: Create a function calc(a, b, operation) where operation can be "add", "sub", "mul", or "div".

def calc(a, b, operation):
    if operation == 'add':
        result = a + b
    elif operation == 'sub':
        result = a - b
    elif operation == 'mul':
        result = a * b
    elif operation == 'div':
        result = a / b
    else:
        print('Enter valid data')
    return result
```

```python
calc(3,4,'add')
```

```python
calc(9,8,'sub')
```

```python
calc(10,4,'mul')
```

```python
calc(8,4,'div')
```

```python
# Q.13 Password Validator: Write a function that checks if a password is longer than 8 characters AND contains the symbol "@".

def pass_validator(password):
    for ch in password:
        if len(password) >= 8 and ch in '@':
            print('Valid password')
            break
    else:
        print('Password is not valid. Password contain @.')
```

```python
pass_validator('renukpatil')
```

```python
# Q.14 Nested Logic: Write a function that takes a list of ages and uses the even_odd logic from your 
# notebook to count how many even-numbered ages exist.

def age_even(ages):
    count = 0
    for i in ages:
        if i % 2 == 0:
            count = count + 1
            print('Even ages: ',i)
    return count
```

```python
age_even([20,15,16,27,28])
```
