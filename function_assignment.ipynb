Q2: Write a program where: 
o You define a global variable counter with an initial value of 0. 
o Create a function increment() that increments counter by 1 each time it is called 
and prints its value. 
o Call the function three times and observe the output.

```python
counter = 0

def increment():
    global counter
    counter = counter + 1
    return counter
```

```python
increment()
```

```python
increment()
```

```python
increment()
```

```python
def calculate_area(radius):
    
    """This function is used to calculate the area of Circle.
    
     Here We used formula :=
     Area = π * r^2
     
     Where, values of π = 22 / 7 
     radius = radius of the circle"""
    
    area = 22 / 7 * (radius * radius)
    return area
```

```python
calculate_area(6)
```

```python
help(calculate_area)
```

```python
# Q7: Write a function is_even(n) that returns True if a number is even, otherwise returns False. Test the function with multiple inputs.

def is_even(n):
    if n % 2 == 0:
        print(n,'is Even number')
    else:
        print(n,'is not Even Number')
```

```python
is_even(49)
```

```python
is_even(6)
```

```python
def pattern6():
    n = 4
    i = 1
    while i <= n:
        print(str(i) * i)
        i = i + 1
```
