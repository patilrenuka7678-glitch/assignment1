```python
# Reverse a String: Input a string and print it in reverse using a for loop.

n = input('Enter the string:')
rev =""
for i in n:
    rev = i + rev
    print(rev)
```

```python
# Q.6 Find Prime Numbers: Print all prime numbers between 1 and a given number N

N = int(input("Enter the number: "))
)
for num in range(2, N+1):   # (2, 3, 4, 5)
    count = 0 # here count is used for count the factor's of given no

    for i in range(1, num+1):   # (1, 2, 3)
        if num % i == 0:     #  num = 2 % 1 == 0 --> True....   
            count = count + 1

    if count == 2:
        print(num)
```

```python
# 8. Sum of Odd and Even Numbers: Take a list of numbers and find the sum of odd and even numbers separately. 

l = [1,2,3,4,5] 
a = 0
b = 0

for i in l:
    if i % 2 == 0:
        a += i
    else:
        b += i
print('Even numbres sum is :',a)
print('Odd numbers sum is :',b)
```
