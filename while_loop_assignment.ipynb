```python
# 4. Reverse Digits: Input a number and reverse its digits using a while loop. Example: 12345 → 54321.

num = 12345  # 1234 #123 # 12
rev = 0  # now rev = 5 now rev = 54  now 543

while num > 0:
    x = num % 10  # 12345 % 10 = 5 so x = 5   # 1234 % 10 = 4 so x = 4  # 123 % 10 = 3 so x = 3
    rev = rev * 10 + x  # 0 * 10 + 5 = 5   # 5 * 10 + 4 = 54    # 54 * 10 + 3 = 543
    num = num // 10   # 12345 // 10 = 1234 # 1234 // 10 = 123 # 123 // 10 = 12
print(rev)
```

```python
# 6. Fibonacci Sequence: Generate the first N terms of the Fibonacci sequence using a while loop.

n = int(input('Enter the number: '))
num1 = 0
num2 = 1

count = 0 # now count = 1 

while count < n:   # 0 < 5:
    print(num1) 
    next_no = num1 + num2  # 0 + 1 = 1
    num1 = num2
    num2 = next_no
    count = count + 1 # 0 + 1 = 1
```

```python
# 7. Check Palindrome Number: Write a program to check whether a number is a palindrome (e.g., 121 is a palindrome). 
```

```python
a = int(input('Enter the number: '))
original = a
rev = 0

while a > 0:
    x = a % 10
    rev = rev * 10 + x
    a = a //10
print(rev)
if rev == original:
        print(original,'is Palindrome number')
else:
        print(original,'is Not palindrome number')
```

```python
# 1. FizzBuzz Program: Print numbers from 1 to 50. For multiples of 3, print "Fizz", for multiples of 5 print "Buzz", 
# and for multiples of both, print "FizzBuzz".

n = int(input('Enter the number: '))

for i in range(1,n+1):
    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```
