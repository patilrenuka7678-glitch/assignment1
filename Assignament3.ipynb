```python
# 8. Union: If s1 = {1, 2} and s2 = {2, 3}, what is s1.union(s2)?
# Solution -->  Its returns the all values of s1 and s2.

s1 = {1,2}
s2 = {2,3}
s1.union(s2)
```

```python
# 9. Intersection: Using the same sets above, what is s1.intersection(s2)?
# Solution --> It's Returns the value which is similar in the both sets.

s1.intersection(s2)
```
