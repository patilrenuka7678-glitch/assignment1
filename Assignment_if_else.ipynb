```python
# 1. Check Eligibility to Vote: Write a program to check if a person is eligible to vote (age >= 18). 
age = int(input('Enter Your Age: '))

if age >= 18:
    print("Congratulation's! Your Eligible for voting")
else:
    print('Sorry! You are not eligible for the voting')
```

```python
# 2. Number Classification: Take an integer input and determine if it is positive, negative, or zero. 
no = int(input('Enter The no: '))

if no > 0:
    print("It's a positive no")
elif no < 0:
    print("It's a Negative no")
else:
    print("It's Zero")
```

```python
marks = float(input('Enter your marks: '))

if marks >= 90 and marks <= 100:
    print('A')
elif marks >= 80 and marks <= 89:
    print('B')
elif marks >= 70 and marks <= 79:
    print('C')
else:
    print('Fail')
```

```python
# 4. Odd or Even: Check whether a given number is even or odd.
num = int(input('Enter the number: '))

if num % 2 == 0:
    print("It's a Even number")
else:
    print("it's a odd number")
```

```python
# 5. Write a program to check whether a year is a leap year or not.

year = int(input('Enter the year: '))

if year % 4 == 0:
    print("yes, It's a leap Year")
else:
    print("No, It's not a leap year")
```
